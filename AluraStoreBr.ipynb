### Importação dos dados



```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_1.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_4.csv"

loja = pd.read_csv(url)
loja2 = pd.read_csv(url2)
loja3 = pd.read_csv(url3)
loja4 = pd.read_csv(url4)

loja.head()
```

#1. Análise do faturamento


```python
# prompt: você deve calcular o faturamento total de cada loja. Somando os valores da coluna Preço de cada loja para estimar o faturamento. E formata para mim os valores com se veem no Brasil

# Para calcular a soma do faturamento
faturamento_loja1 = loja['Preço'].sum()
faturamento_loja2 = loja2['Preço'].sum()
faturamento_loja3 = loja3['Preço'].sum()
faturamento_loja4 = loja4['Preço'].sum()

# Fromato de dinherio brasileiro
faturamento_loja1_br = f"R$ {faturamento_loja1:,.2f}".replace(",", "X").replace(".", ",").replace("X", ".")
faturamento_loja2_br = f"R$ {faturamento_loja2:,.2f}".replace(",", "X").replace(".", ",").replace("X", ".")
faturamento_loja3_br = f"R$ {faturamento_loja3:,.2f}".replace(",", "X").replace(".", ",").replace("X", ".")
faturamento_loja4_br = f"R$ {faturamento_loja4:,.2f}".replace(",", "X").replace(".", ",").replace("X", ".")

print(f"Faturamento da Loja 1: {faturamento_loja1_br}")
print(f"Faturamento da Loja 2: {faturamento_loja2_br}")
print(f"Faturamento da Loja 3: {faturamento_loja3_br}")
print(f"Faturamento da Loja 4: {faturamento_loja4_br}")
```

# 2. Vendas por Categoria


```python
# prompt: deve calcular a quantidade de produtos vendidos por categoria em cada loja. A ideia é agrupar os dados por categoria e contar o número de vendas de cada tipo, mostrando as categoria mais populares de cada loja

# Vendas por categoria em cada loja
vendas_por_categoria_loja1 = loja.groupby('Categoria do Produto')['Categoria do Produto'].count().sort_values(ascending=False)
vendas_por_categoria_loja2 = loja2.groupby('Categoria do Produto')['Categoria do Produto'].count().sort_values(ascending=False)
vendas_por_categoria_loja3 = loja3.groupby('Categoria do Produto')['Categoria do Produto'].count().sort_values(ascending=False)
vendas_por_categoria_loja4 = loja4.groupby('Categoria do Produto')['Categoria do Produto'].count().sort_values(ascending=False)

print("Vendas por Categoria - Loja 1:")
print(vendas_por_categoria_loja1)
print("\nVendas por Categoria - Loja 2:")
print(vendas_por_categoria_loja2)
print("\nVendas por Categoria - Loja 3:")
print(vendas_por_categoria_loja3)
print("\nVendas por Categoria - Loja 4:")
print(vendas_por_categoria_loja4)
```

# 3. Média de Avaliação das Lojas

```python
# prompt: Neste passo, vamos calcular a média das avaliações dos clientes para cada loja. O objetivo é entender a satisfação dos clientes com os produtos vendidos.

# Cálculo da média de avaliação para cada loja
media_avaliacao_loja1 = loja['Avaliação da compra'].mean()
media_avaliacao_loja2 = loja2['Avaliação da compra'].mean()
media_avaliacao_loja3 = loja3['Avaliação da compra'].mean()
media_avaliacao_loja4 = loja4['Avaliação da compra'].mean()

# Imprimir as médias de avaliação formatadas
print(f"Média de avaliação da Loja 1: {media_avaliacao_loja1:.2f}")
print(f"Média de avaliação da Loja 2: {media_avaliacao_loja2:.2f}")
print(f"Média de avaliação da Loja 3: {media_avaliacao_loja3:.2f}")
print(f"Média de avaliação da Loja 4: {media_avaliacao_loja4:.2f}")
```

# 4. Produtos Mais e Menos Vendidos

```python
sns.set(style="whitegrid")

# Calcular os produtos mais e menos vendidos por loja
produtos_mais_vendidos_loja1 = loja['Produto'].value_counts().nlargest(5)
produtos_menos_vendidos_loja1 = loja['Produto'].value_counts().nsmallest(5)

produtos_mais_vendidos_loja2 = loja2['Produto'].value_counts().nlargest(5)
produtos_menos_vendidos_loja2 = loja2['Produto'].value_counts().nsmallest(5)

produtos_mais_vendidos_loja3 = loja3['Produto'].value_counts().nlargest(5)
produtos_menos_vendidos_loja3 = loja3['Produto'].value_counts().nsmallest(5)

produtos_mais_vendidos_loja4 = loja4['Produto'].value_counts().nlargest(5)
produtos_menos_vendidos_loja4 = loja4['Produto'].value_counts().nsmallest(5)

# Criar os subplots
fig, axes = plt.subplots(4, 2, figsize=(18, 22))

# Lista de dados para iteração
dados = [
    (produtos_mais_vendidos_loja1, "Mais Vendidos - Loja 1", axes[0, 0], "#4CAF50"),
    (produtos_menos_vendidos_loja1, "Menos Vendidos - Loja 1", axes[0, 1], "#F44336"),

    (produtos_mais_vendidos_loja2, "Mais Vendidos - Loja 2", axes[1, 0], "#4CAF50"),
    (produtos_menos_vendidos_loja2, "Menos Vendidos - Loja 2", axes[1, 1], "#F44336"),

    (produtos_mais_vendidos_loja3, "Mais Vendidos - Loja 3", axes[2, 0], "#4CAF50"),
    (produtos_menos_vendidos_loja3, "Menos Vendidos - Loja 3", axes[2, 1], "#F44336"),

    (produtos_mais_vendidos_loja4, "Mais Vendidos - Loja 4", axes[3, 0], "#4CAF50"),
    (produtos_menos_vendidos_loja4, "Menos Vendidos - Loja 4", axes[3, 1], "#F44336")
]

# Criar os gráficos
for valores, titulo, ax, cor in dados:
    df_plot = pd.DataFrame({'Produto': valores.index, 'Quantidade': valores.values})
    sns.barplot(data=df_plot, x='Produto', y='Quantidade', hue='Produto', palette={p: cor for p in df_plot['Produto']}, ax=ax, legend=False)
    ax.set_title(titulo, fontsize=14, fontweight='bold')
    ax.set_xlabel("Produto", fontsize=12)
    ax.set_ylabel("Quantidade", fontsize=12)
    ax.tick_params(axis='x', rotation=45)

# Ajuste final
plt.suptitle("Produtos Mais e Menos Vendidos por Loja", fontsize=18, fontweight='bold', y=1.02)
plt.tight_layout()
plt.show()
```


# 5. Frete Médio por Loja


```python
# Estilo visual
sns.set(style="whitegrid")

# 1. Cálculo do frete médio por loja
frete_medio_loja1 = loja['Frete'].mean()
frete_medio_loja2 = loja2['Frete'].mean()
frete_medio_loja3 = loja3['Frete'].mean()
frete_medio_loja4 = loja4['Frete'].mean()

# Impressão do frete médio por loja
print(f"Frete médio da Loja 1: R$ {frete_medio_loja1:.2f}")
print(f"Frete médio da Loja 2: R$ {frete_medio_loja2:.2f}")
print(f"Frete médio da Loja 3: R$ {frete_medio_loja3:.2f}")
print(f"Frete médio da Loja 4: R$ {frete_medio_loja4:.2f}")

# 2. Cálculo do impacto do frete médio sobre o preço médio
impacto_frete_medio_loja1 = (loja['Frete'].mean() / loja['Preço'].mean()) * 100
impacto_frete_medio_loja2 = (loja2['Frete'].mean() / loja2['Preço'].mean()) * 100
impacto_frete_medio_loja3 = (loja3['Frete'].mean() / loja3['Preço'].mean()) * 100
impacto_frete_medio_loja4 = (loja4['Frete'].mean() / loja4['Preço'].mean()) * 100

# Impressão do impacto médio do frete
print(f"Impacto médio do frete no preço da Loja 1: {impacto_frete_medio_loja1:.2f}%")
print(f"Impacto médio do frete no preço da Loja 2: {impacto_frete_medio_loja2:.2f}%")
print(f"Impacto médio do frete no preço da Loja 3: {impacto_frete_medio_loja3:.2f}%")
print(f"Impacto médio do frete no preço da Loja 4: {impacto_frete_medio_loja4:.2f}%")

# 3. Visualização em gráfico de colunas
labels = ['Loja 1', 'Loja 2', 'Loja 3', 'Loja 4']
impacto_frete_medio = [
    impacto_frete_medio_loja1,
    impacto_frete_medio_loja2,
    impacto_frete_medio_loja3,
    impacto_frete_medio_loja4
]

# Construir DataFrame auxiliar para evitar warnings do seaborn
df_plot = pd.DataFrame({
    'Loja': labels,
    'Impacto Médio (%)': impacto_frete_medio
})

plt.figure(figsize=(8, 6))
sns.barplot(data=df_plot, x='Loja', y='Impacto Médio (%)', hue='Loja', palette='coolwarm', dodge=False, legend=False)

plt.title('Impacto do Frete Médio no Preço Médio por Loja', fontsize=14, fontweight='bold')
plt.ylabel('Impacto Médio (%)')
plt.xlabel('Loja')
plt.ylim(0, max(impacto_frete_medio) + 5)

# Mostrar o valor em cima de cada barra
for i, v in enumerate(impacto_frete_medio):
    plt.text(i, v + 0.5, f"{v:.2f}%", ha='center', va='bottom', fontweight='bold')

plt.tight_layout()
plt.show()
```
